```python
!pip install matplotlib pandas seaborn --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de producción de arándanos (2015–2024)
data = {
    'Año': list(range(2015, 2025)),
    'Huelva (España)': [48000, 51000, 54000, 56000, 60000, 62000, 63000, 65000, 66000, 67000],
    'Perú': [30000, 50000, 70000, 95000, 120000, 160000, 190000, 230000, 270000, 300000],
    'EE.UU.': [275000, 278000, 278500, 278000, 277000, 275000, 274000, 273000, 273000, 272000],
    'Chile': [100000, 110000, 120000, 130000, 140000, 145000, 150000, 153000, 155000, 157000],
    'Marruecos': [12000, 15000, 19000, 23000, 28000, 33000, 40000, 47000, 52000, 55000],
    'México': [25000, 30000, 36000, 42000, 48000, 55000, 60000, 64000, 67000, 69000]
}

df = pd.DataFrame(data)

df_long = df.melt(id_vars='Año', var_name='País', value_name='Producción')

colors = {
    'Huelva (España)': '#1f77b4',
    'Perú': '#d62728',
    'EE.UU.': '#2ca02c',
    'Chile': '#ff7f0e',
    'Marruecos': '#9467bd',
    'México': '#8c564b'
}

sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
sns.barplot(data=df_long, x="Año", y="Producción", hue="País", palette=colors, ci=None)

plt.title("Producción de Arándanos por País (2015–2024)", fontsize=18, weight='bold')
plt.xlabel("Año", fontsize=12)
plt.ylabel("Toneladas métricas", fontsize=12)
plt.legend(title="País")
plt.tight_layout()
plt.show()
```
