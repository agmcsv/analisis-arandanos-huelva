```python
!pip install matplotlib pandas seaborn --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos aproximados de producción anual de arándanos por país en toneladas métricas
data = {
    'Año': list(range(2015, 2025)),
    'Huelva (España)': [48000, 51000, 54000, 56000, 60000, 62000, 63000, 65000, 66000, 67000],
    'Perú': [30000, 50000, 70000, 95000, 120000, 160000, 190000, 230000, 270000, 300000],
    'EE.UU.': [275000, 278000, 278500, 278000, 277000, 275000, 274000, 273000, 273000, 272000],
    'Chile': [100000, 110000, 120000, 130000, 140000, 145000, 150000, 153000, 155000, 157000],
    'Marruecos': [12000, 15000, 19000, 23000, 28000, 33000, 40000, 47000, 52000, 55000],
    'México': [25000, 30000, 36000, 42000, 48000, 55000, 60000, 64000, 67000, 69000]
}

df = pd.DataFrame(data)

# Paleta personalizada de colores bien contrastados
custom_colors = {
    'Huelva (España)': '#1f77b4',  # Azul fuerte
    'Perú': '#d62728',             # Rojo
    'EE.UU.': '#2ca02c',           # Verde
    'Chile': '#ff7f0e',            # Naranja
    'Marruecos': '#9467bd',        # Púrpura
    'México': '#8c564b'            # Marrón
}

sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

for pais in df.columns[1:]:
    plt.plot(df['Año'], df[pais], label=pais, linewidth=2.8, marker='o', color=custom_colors[pais])

plt.title("Comparativa de Producción de Arándanos (2015–2024)", fontsize=18, weight='bold')
plt.xlabel("Año", fontsize=12)
plt.ylabel("Producción (toneladas métricas)", fontsize=12)
plt.xticks(df['Año'])
plt.legend(title="País", fontsize=10)
plt.tight_layout()
plt.show()
```
