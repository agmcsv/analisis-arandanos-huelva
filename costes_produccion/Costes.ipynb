```python
import pandas as pd

url = "https://raw.githubusercontent.com/agmcsv/analisis-arandanos-huelva/main/costes_produccion/costes_chile.csv"
df = pd.read_csv(url, encoding='utf-8')
df.columns = df.columns.str.strip()

print("Columnas:", df.columns.tolist())
df.head()
```
