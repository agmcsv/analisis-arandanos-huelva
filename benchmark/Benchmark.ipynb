```python
import pandas as pd
import requests
from io import StringIO

# URL del archivo .md en GitHub (raw)
url_md = "https://raw.githubusercontent.com/agmcsv/analisis-arandanos-huelva/main/benchmark/comparativa_huelva_vs_marruecos.md"

# Descargar contenido
response = requests.get(url_md)
md_text = response.text

# Extraer líneas que contienen tabla Markdown (líneas con '|')
table_lines = [line for line in md_text.split('\n') if '|' in line]

# Filtrar líneas separadoras (líneas con guiones)
table_lines = [line for line in table_lines if not all(c in '-| :' for c in line)]

# Construir string de tabla para pandas
table_str = '\n'.join(table_lines)

# Leer tabla con pandas (sep='|', quitar columnas vacías por bordes)
df = pd.read_csv(StringIO(table_str), sep='|').dropna(axis=1, how='all')

# Limpiar espacios de columnas y filas
df.columns = df.columns.str.strip()
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df.head()
```
